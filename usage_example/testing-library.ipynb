# Installing libraries

```python
from bordersandpoints import border_lines_cap,create_video,process_video
```

```python
import os
video_file = ["1710074366570.mp4"] # list of video file names
video_path = "/kaggle/input/check-check" # path to the video file
os.mkdir("/kaggle/working/testing") # create a folder to save the output
output_path = "/kaggle/working/testing" # path to save the output file
```

# Tracking point in the whole trunk video

```python
initial_point=(500, 600) # initial point to track in the whole video
process_video(video_file,video_path,output_path,initial_point) # tracking the initial point in the whole video using farneback algorithm
```

# Drawing border lines to trunk in the whole video

```python
lines_output = "/kaggle/working/testing/videos" # path to save the output image frames of input video which contains the border lines
os.mkdir(lines_output) # create a folder to save the output
border_lines_cap(video_file,output_path, lines_output) # draw border lines to every frame in the video and outputs the frames with border lines
```
